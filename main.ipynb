```python
from diagrams import Cluster, Diagram, Edge
from diagrams.custom import Custom
from diagrams.programming.framework import Angular
from diagrams.onprem.database import Mongodb
from diagrams.azure.web import APIConnections
from diagrams.programming.framework import FastAPI 
from diagrams.onprem.database import Postgresql
from diagrams.gcp.ml import TTS
from diagrams.gcp.ml import STT
from diagrams.programming.framework import React

graph_attr = {
    "splines":"curved"
    }

with Diagram(name="Abair ecosystem", show=False, graph_attr=graph_attr) as diag:
    
    with Cluster("Abair Server"):
    
        with Cluster("APIs"):
            with Cluster("Synthesis"):
                synthesis = TTS("Synthesis")
                TTSApi = FastAPI('FastAPI')
                
            with Cluster("Recognition"):
                recognition = STT("Recognition")
            
        with Cluster("Applications"):   
            
            with Cluster("An Scéalaí"):
                ASdatabase = Mongodb("MongoDB")

                with Cluster("AS Node"):
                    ASfrontend = Angular("frontend")

                    with Cluster("Express"):
                        ASexpress = Custom("Express", "./icons/express.png")
                        ASauth = Custom("Passport", "./icons/passport.png")

            with Cluster("Leon Don Lón"):
                LDLdatabase = Postgresql("Postgres")
                
                with Cluster("LDL Node"):
                    LDLfrontend = React("frontend")
                    LDLexpress = Custom("Express", "./icons/express.png")
                    LDLbackend = Custom("Supabase", "./icons/supabase.png")
        
    ASexpress >> ASfrontend
    ASdatabase >> ASexpress
    TTSApi >> ASexpress
    
    LDLexpress >> LDLfrontend
    LDLdatabase >> LDLexpress
    
diag
```
